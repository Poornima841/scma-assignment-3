```python
import pandas as pd

# Load the dataset
data = pd.read_csv("path_to_your_dataset.csv")

python
Copy code
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_curve, auc, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Define the features and target
X = data[['feature1', 'feature2', 'feature3']]  # Replace with your features
y = data['target']  # Replace with your target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Add constant term for the intercept
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

# Fit the logistic regression model
logit_model = sm.Logit(y_train, X_train).fit()

# Summary of the logistic regression model
print(logit_model.summary())

# Predict probabilities
y_pred_prob = logit_model.predict(X_test)

# Predict binary outcomes
y_pred = (y_pred_prob > 0.5).astype(int)

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)

# ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
Step 3: Decision Tree Analysis
python
Copy code
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

# Fit the decision tree model
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Predict with the decision tree model
y_pred_dt = dt_model.predict(X_test)

# Confusion matrix for decision tree
conf_matrix_dt = confusion_matrix(y_test, y_pred_dt)
print(conf_matrix_dt)

# ROC curve and AUC for decision tree
y_pred_prob_dt = dt_model.predict_proba(X_test)[:, 1]
fpr_dt, tpr_dt, thresholds_dt = roc_curve(y_test, y_pred_prob_dt)
roc_auc_dt = auc(fpr_dt, tpr_dt)

# Plot ROC curve for decision tree
plt.figure()
plt.plot(fpr_dt, tpr_dt, color='green', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_dt)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
Part B: Probit Regression Analysis
python
Copy code
# Probit regression analysis
probit_model = sm.Probit(y_train, X_train).fit()

# Summary of the probit regression model
print(probit_model.summary())

# Predict probabilities using the probit model
y_pred_prob_probit = probit_model.predict(X_test)

# Confusion matrix for probit regression
y_pred_probit = (y_pred_prob_probit > 0.5).astype(int)
conf_matrix_probit = confusion_matrix(y_test, y_pred_probit)
print(conf_matrix_probit)

# ROC curve and AUC for probit regression
fpr_probit, tpr_probit, thresholds_probit = roc_curve(y_test, y_pred_prob_probit)
roc_auc_probit = auc(fpr_probit, tpr_probit)

# Plot ROC curve for probit regression
plt.figure()
plt.plot(fpr_probit, tpr_probit, color='red', lw=2, label='ROC curve (area = %0.2f)' % roc_auc_probit)
plt.plot([0, 1], [0, 1], color='gray', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()

python
Copy code
from statsmodels.discrete.discrete_model import Tobit

# Define the dependent variable and censoring limits
# Assuming 'target' is the dependent variable and it's left-censored at 0 and right-censored at a certain value
dependent_var = data['target']
censored_lower_limit = 0
censored_upper_limit = dependent_var.max()  # Adjust based on your data

# Fit the Tobit model
tobit_model = Tobit(dependent_var, X_train, left=censored_lower_limit, right=censored_upper_limit).fit()

# Summary of the Tobit model
print(tobit_model.summary())

# Predict using the Tobit model
y_pred_tobit = tobit_model.predict(X_test)

# Compare predicted values with actual values
plt.figure()
plt.scatter(y_test, y_pred_tobit)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values (Tobit Model)')
plt.show()
```
